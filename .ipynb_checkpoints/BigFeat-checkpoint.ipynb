```python
baseDir=r'C:\Users\kayak\Downloads\BigFeat-master\BigFeat-master\bigfeat\AutoMl_datasets\New folder'
datasets=os.listdir(baseDir)
ml_models={"Logistic_Regression":LogisticRegression(),
           "AdaBoostClassifier":AdaBoostClassifier(n_estimators=100, random_state=0),
           "DecisionTreeClassifier":DecisionTreeClassifier(random_state=0, max_depth=5),
           "ExtraTreesClassifier":ExtraTreesClassifier(n_estimators=100, random_state=0),
           "KNNClassifier":KNeighborsClassifier(n_neighbors=3),
           "MLP":MLPClassifier(random_state=1, max_iter=300),
           "RandomForestClassifier":RandomForestClassifier(max_depth=2, random_state=0),
           "SVM":svm.SVC(kernel='linear'),
           "GradientBoostingClassifier":GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0)
          }
    
```

```python

iterables = [[i.split('.')[0] for i in datasets], ["Original", "Generated"]]

index = pd.MultiIndex.from_product(iterables, names=["DataSet", "Type"])

dataframe= pd.DataFrame(columns=[i for i in ml_models]+['Framework'], index=index)
    
dataframe
```

```python
def getData(directory):
    data=pd.read_csv(directory)
    y=data.Label
    X=data.drop('Label',axis=1)
    return X,y
```

```python
def mlPipeline (X,y,ml_models,data_name,Framework='BigFeat'):
    
    if Framework=='BigFeat':
    
        feature_generation=BigFeat()
        generated_data=feature_generation.fit(X,y)
    
        for i in ml_models:
        
            #originial Feature
        
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
            model=ml_models[i]
            model.fit(X_train,y_train)
            y_pred=model.predict(X_test)
            score=roc_auc_score(y_pred,y_test)
            dataframe.loc[(data_name,"Original")][i]=score
            dataframe.loc[(data_name,"Original")]['Framework']=Framework
        
        
        
        #Generated Feature
        
            X_train, X_test, y_train, y_test = train_test_split(generated_data, y, test_size=0.33, random_state=42)
            model=ml_models[i]
            model.fit(X_train,y_train)
            y_pred=model.predict(X_test)
            score=roc_auc_score(y_pred,y_test)
            dataframe.loc[(data_name,"Generated")][i]=score
            dataframe.loc[(data_name,"Generated")]['Framework']=Framework
        
        
        
        
```

```python
def startPipeline(directory,datasets,ml_models):
    
    for dataset in datasets:
        X,y=getData(f'{directory}/{dataset}')
        
        mlPipeline(X,y,ml_models,dataset.split('.')[0])
    return dataframe
```

```python
startPipeline(baseDir,datasets,ml_models)
```
